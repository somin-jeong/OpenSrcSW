```python
import csv
import matplotlib.pyplot as plt
f = open('C:\seoulw.csv')
data = csv.reader(f)
header = next(data)
resultmax = []
resultmin = []
for row in data:
    if row[-1] != '':
        if row[0].split('-')[1] == '11' and row[0].split('-')[2] == '04':
            resultmax.append(float(row[-1]))
    if row[-2] != '':
        if row[0].split('-')[1] == '11' and row[0].split('-')[2] == '04':
            resultmin.append(float(row[-2]))       

plt.plot(resultmax, color='deeppink')
plt.plot(resultmin, color='skyblue')
plt.show()

```

```python
import csv
import matplotlib.pyplot as plt
f = open('C:\seoulw.csv')
data = csv.reader(f)
header = next(data)
aug = []
jan = []

for row in data:
    month = row[0].split('-')[1]
    if row[-1] != '':
        if month == '08':
            aug.append(float(row[-1]))
        if month == '01':
            jan.append(float(row[-1]))
            
plt.hist(aug, bins=100, color='r', label='Aug')
plt.hist(jan, bins=100, color='b', label='Jan')
plt.legend()
plt.show()
```
